```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Load dataset
file_path = 'Final Data2.csv'
data = pd.read_csv(file_path)

# Label encode categorical features 'District' and 'Items'
label_encoders = {}
for col in ['District', 'Items']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Define weather-related columns
weather_columns = ['Root_Zone_Soil_Wetness', 'Forst Point 2 meter Temp',
                    'Wind speed 2 meters', 'Average Solar Radiation',
                   'Average Rainfall (mm)', 'Average Sunshine (Hours)', 'Average Cloud Coverage (Octs)',
                   'Average pet', 'Average Humidity (percent)', 'Average_Max_Temperature (Deg.Cel)',
                   'Average Min Temperature (Deg.Cel)']

# Initialize MinMaxScaler for weather and production data
scaler_weather = MinMaxScaler()
scaler_production = MinMaxScaler()

# Normalize weather data
data[weather_columns] = scaler_weather.fit_transform(data[weather_columns])

# Normalize production data
data['production'] = scaler_production.fit_transform(data['production'].values.reshape(-1, 1))

# Group data by district and year
data_grouped = data.groupby(['District','Year','Items']).mean().reset_index()

# Define function to create sequences for CNN
def create_sequences(data, time_steps=6):
    sequences, labels = [], []
    for i in range(len(data) - time_steps):
        sequence = data.iloc[i:i + time_steps].values
        label = data.iloc[i + time_steps].values
        sequences.append(sequence)
        labels.append(label)
    return np.array(sequences), np.array(labels)

# Prepare the weather data sequences
weather_sequences, weather_labels = create_sequences(data_grouped[weather_columns], time_steps=4)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(weather_sequences, weather_labels, test_size=0.2, random_state=42)

# Build 1D-CNN model for weather prediction
model_cnn_weather = Sequential([
    Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),
    Dropout(0.1),
    Conv1D(filters=128, kernel_size=1, activation='relu'),  # Changed kernel size to 1 to prevent shrinking too much
    Flatten(),
    Dense(64, activation='relu'),
    Dense(y_train.shape[1])
])

# Compile the model with a custom learning rate
optimizer_weather_cnn = Adam(learning_rate=0.006167)
model_cnn_weather.compile(optimizer=optimizer_weather_cnn, loss='mse')

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the CNN model for weather prediction
history_cnn_weather = model_cnn_weather.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=80, batch_size=32, callbacks=[early_stopping])

# Predict the weather sequences using the trained CNN model
weather_predictions_full_cnn = model_cnn_weather.predict(weather_sequences)

# Ensure the predicted weather data has the correct shape and aligns with the data index
data_grouped_predicted_cnn = data_grouped.copy()

# Replace the last time step of the weather data with the predicted values
for i, col in enumerate(weather_columns):
    data_grouped_predicted_cnn[col].iloc[-len(weather_predictions_full_cnn):] = weather_predictions_full_cnn[:, i]

# Use predicted weather data for production prediction
production_columns_predicted = ['production'] + weather_columns
production_sequences_predicted_cnn, production_labels_predicted_cnn = create_sequences(data_grouped_predicted_cnn[production_columns_predicted], time_steps=4)

# Split the data into training and testing sets for production prediction using CNN
X_train_prod_pred_cnn, X_test_prod_pred_cnn, y_train_prod_pred_cnn, y_test_prod_pred_cnn = train_test_split(production_sequences_predicted_cnn, production_labels_predicted_cnn, test_size=0.2, random_state=42)

# Build 1D-CNN model for production prediction using predicted weather data
model_cnn_prod_pred = Sequential([
    Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train_prod_pred_cnn.shape[1], X_train_prod_pred_cnn.shape[2])),
    MaxPooling1D(pool_size=2),
    Dropout(0.1),
    Conv1D(filters=128, kernel_size=1, activation='relu'),  # Adjusted kernel size to 1 to prevent shrinking
    Flatten(),
    Dense(64, activation='relu'),
    Dense(y_train_prod_pred_cnn.shape[1])
])

# Compile the production prediction CNN model
optimizer_prod_pred_cnn = Adam(learning_rate=0.006167)
model_cnn_prod_pred.compile(optimizer=optimizer_prod_pred_cnn, loss='mse')

# Train the CNN model for production prediction
history_prod_pred_cnn = model_cnn_prod_pred.fit(X_train_prod_pred_cnn, y_train_prod_pred_cnn, validation_data=(X_test_prod_pred_cnn, y_test_prod_pred_cnn), epochs=100, batch_size=32, callbacks=[early_stopping])

# Predict production data using the CNN model
production_predictions_pred_cnn = model_cnn_prod_pred.predict(X_test_prod_pred_cnn)

# Convert predictions back to original scale
production_predictions_pred_rescaled_cnn = scaler_production.inverse_transform(production_predictions_pred_cnn)

# Calculate metrics
production_mse_cnn = mean_squared_error(y_test_prod_pred_cnn, production_predictions_pred_cnn)
production_rmse_cnn = np.sqrt(production_mse_cnn)
production_mae_cnn = mean_absolute_error(y_test_prod_pred_cnn, production_predictions_pred_cnn)
production_r2_cnn = r2_score(y_test_prod_pred_cnn, production_predictions_pred_cnn)
production_mape_cnn = mean_absolute_percentage_error(y_test_prod_pred_cnn, production_predictions_pred_cnn)

threshold = 0.1
production_accuracy_cnn = np.mean(np.abs((y_test_prod_pred_cnn - production_predictions_pred_cnn) / y_test_prod_pred_cnn) < threshold) * 100

# Print metrics for CNN model
print(f"Production Prediction CNN - MSE: {production_mse_cnn}, RMSE: {production_rmse_cnn}, MAE: {production_mae_cnn}, R²: {production_r2_cnn}, MAPE: {production_mape_cnn}")
print(f"Production Prediction CNN Custom Accuracy (within 10% threshold): {production_accuracy_cnn}%")

# Plot loss for CNN weather prediction
plt.plot(history_cnn_weather.history['loss'])
plt.plot(history_cnn_weather.history['val_loss'])
plt.title('Model loss - CNN Weather Prediction')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()

# Plot loss for CNN production prediction
plt.plot(history_prod_pred_cnn.history['loss'])
plt.plot(history_prod_pred_cnn.history['val_loss'])
plt.title('Model loss - CNN Production Prediction')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```
